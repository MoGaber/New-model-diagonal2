```python
import copy
class vertex:
    def __init__(self, val):
        self.position = val
        #
        self.right_neighbor = None
        self.left_neighbor = None
        self.top_neighbor = None
        self.bottom_neighbor = None
        #
        self.right_state = "o"
        self.left_state = "o"
        self.top_state = "o"
        self.bottom_state = "o" 

        

def generate_lattice(L):
    global vertices_list
    vertices_list = [ []*i for i in range(L+1)]
    for row in range(L+1):
        for column in range(L+1):
            vertices_list[row].append(str([row+1, column+1]))
            vertices_list[row][-1] = vertex([row+1, column+1])

    for row in range(1, L+2):
        for column in range(1, L+2):
                if column < L+1:
                    set_horizontal_neighbor(call_vertex(row, column), call_vertex(row, column+1))
                if row >1:
                    set_vertical_neighbor(call_vertex(row, column), call_vertex(row-1, column))


                if row ==L+1 :
                    vertical_equalize(call_vertex(row, column), call_vertex(row-L, column))

        horizontal_equalize(call_vertex(row, 1), call_vertex(row, L+1))
#########    
                

def set_horizontal_neighbor(vertex1, vertex2):
        vertex1.right_neighbor=vertex2
        vertex2.left_neighbor = vertex1
    
def set_vertical_neighbor(vertex1, vertex2):
        vertex1.top_neighbor=vertex2
        vertex2.bottom_neighbor = vertex1


def call_vertex(x, y):
    global vertices_list
    return vertices_list[x-1][y-1]

def horizontal_equalize(vertex1, vertex2):
    vertex1.left_neighbor = vertex2.left_neighbor
    vertex2.right_neighbor = vertex1.right_neighbor
    
    
def vertical_equalize(vertex1, vertex2):
    vertex2.top_neighbor = vertex1.top_neighbor
    vertex1.bottom_neighbor = vertex2.bottom_neighbor 
    
```

```python
state_1 = vertex([1, 1])
state_1.right_state = '→'
state_1.left_state = '→'
state_1.top_state = '↑'
state_1.bottom_state = '↑'
####
state_2 = vertex([1, 1])
state_2.right_state = '→'
state_2.left_state = '→'
state_2.top_state = '↓'
state_2.bottom_state = '↓'
#####
state_3 = vertex([1, 1])
state_3.right_state = '←'
state_3.left_state = '←'
state_3.top_state = '↑'
state_3.bottom_state = '↑'
#####
state_4 = vertex([1, 1])
state_4.right_state = '←'
state_4.left_state = '←'
state_4.top_state = '↓'
state_4.bottom_state = '↓'
#####
state_5 = vertex([1, 1])
state_5.right_state = '←'
state_5.left_state = '→'
state_5.top_state = '↑'
state_5.bottom_state = '↓'
#####
state_6 = vertex([1, 1])
state_6.right_state = '→'
state_6.left_state = '←'
state_6.top_state = '↓'
state_6.bottom_state = '↑'
######

six_states = [state_1, state_2,state_3,state_4,state_5,state_6]

def Gauss_value(node):
    if node.right_state == "o" or node.left_state == 'o' or node.top_state =='o' or node.bottom_state =='o':
        return f'Node {node.position} is not complete'
    value = 0
    if node.right_state == '→':
        value+= -0.5
    elif node.right_state == '←' :
        value+= 0.5
   
    if node.left_state == '→':
        value+= 0.5
    elif node.left_state == '←' :
        value+= -0.5
        
    if node.top_state == '↑':
        value+= -0.5
    elif node.top_state == '↓' :
        value+= 0.5 
    
    if node.bottom_state == '↑':
        value+= 0.5
    elif node.bottom_state == '↓' :
        value+= -0.5 
    return value

def assign_state(vertex, state):
    vertex.right_state = state.right_state
    vertex.left_state = state.left_state
    vertex.top_state = state.top_state
    vertex.bottom_state = state.bottom_state
    #
    vertex.right_neighbor.left_state = vertex.right_state
    vertex.left_neighbor.right_state = vertex.left_state
    vertex.top_neighbor.bottom_state = vertex.top_state
    vertex.bottom_neighbor.top_state = vertex.bottom_state

    
    
def print_vertex(vertex):
    a = vertex.left_state 
    b = vertex.top_state 
    c = vertex.bottom_state 
    d = vertex.right_state 
    
    print(a, end = '')
    print(" "+b, '\n  '+c , ' '+d,end = '')
    
def vertex_equalizer(vertex1, vertex2):
    if vertex1.right_state == 'o':
        vertex1.right_state = vertex2.right_state
    if vertex1.left_state == 'o':
        vertex1.left_state = vertex2.left_state
    if vertex1.top_state == 'o':
        vertex1.top_state = vertex2.top_state
    if vertex1.bottom_state == 'o':
        vertex1.bottom_state = vertex2.bottom_state
    ##
    if vertex2.right_state == 'o':
        vertex2.right_state = vertex1.right_state
    if vertex2.left_state == 'o':
        vertex2.left_state = vertex1.left_state
    if vertex2.top_state == 'o':
        vertex2.top_state = vertex1.top_state
    if vertex2.bottom_state == 'o':
        vertex2.bottom_state = vertex1.bottom_state
        
def periodic_equalizer(vertices_list):
    for i in range(len(vertices_list[0])):
        vertex_equalizer(vertices_list[0][i], vertices_list[-1][i])
        vertex_equalizer(vertices_list[i][0], vertices_list[i][-1])

def naive_print(vertices_list):
    for j in vertices_list:
        print('row')
        for i in j:
            print_vertex(i)
            print('\n')

def check_state(vertices_list):
    for j in vertices_list:
        for i in j:
            if Gauss_value(i) != 0:
                #print(f"vertex {i.position} not allowed")
                return False
    return True

def test():
    counter=0
    for i in six_states:
        for j in six_states:
            generate_lattice(2)
            assign_state(call_vertex(1, 1),i)
            assign_state(call_vertex(2,2), j)
            periodic_equalizer(vertices_list)
            if check_state(vertices_list):
                counter+=1
    return counter
                
def incomplete(node):
    if node.right_state != "o" or node.left_state != 'o' or node.top_state !='o' or node.bottom_state !='o':
        if node.right_state == "o" or node.left_state == 'o' or node.top_state =='o' or node.bottom_state =='o':
            return True
    return False

def align(vertex, state):
    n=0
    if vertex.right_state != 'o':
        if vertex.right_state == state.right_state:
            n+=1
    if vertex.left_state != 'o':
        if vertex.left_state == state.left_state:
            n+=1
    
    if vertex.top_state != 'o':
        if vertex.top_state == state.top_state:
            n+=1        
    
    if vertex.bottom_state != 'o':
        if vertex.bottom_state == state.bottom_state:
            n+=1         
    if n>=2:
        return True
    else:
        return False
    
def good_print(vertices_list):

    for num, row in enumerate(vertices_list):
            print('Row'+str(num+1)+':\n')
            for i in range(3):
                for j in range(len(row)):
                    ll = [row[j].top_state, row[j].left_state, row[j].right_state, row[j].bottom_state]
                    if i ==0:
                        print('  ', end='')
                        print(ll[i], end='')
                        print('  ', end='')
                        if j == len(row)-1:
                            print('  '*6)
                        else:
                            print('  '*6, end='')

                    elif i == 1:
                        print(ll[i], end='')
                        print('  '*3, end='')
                        print(ll[i+1], end='')
                        if j == len(row)-1:
                            print('  '*4)
                        else:
                            print('  '*4, end='')

                    elif i ==2:
                        print('  ', end='')
                        print(ll[i+1], end='')
                        print('  ', end='')
                        if j == len(row)-1:
                            print('  '*6)
                        else:
                            print('  '*6, end='')

            print('\n\n\n')
            
def one_missing(node):
    n=0
    if node.right_state != 'o':
            n+=1
    if node.left_state != 'o':
            n+=1
    if node.top_state != 'o':
            n+=1    
    if node.bottom_state != 'o':
            n+=1 
    if n==3:
        return True
    else:
        return False
    
def align_3(vertex, state):
    n=0
    if vertex.right_state != 'o':
        if vertex.right_state == state.right_state:
            n+=1
    if vertex.left_state != 'o':
        if vertex.left_state == state.left_state:
            n+=1
    
    if vertex.top_state != 'o':
        if vertex.top_state == state.top_state:
            n+=1        
    
    if vertex.bottom_state != 'o':
        if vertex.bottom_state == state.bottom_state:
            n+=1         
    if n==3:
        return True
    else:
        return False


def complete_states(vertices_list):
        for row in vertices_list:
            for verto in row:
                if one_missing(verto):
                    for state in six_states:
                            if align_3(verto, state):
                                assign_state(verto, state)
                                periodic_equalizer(vertices_list)
                                
                                
def test_3():

    counter=0
    for i in six_states:
        for j in six_states:
            for b in six_states:
                generate_lattice(3)
                assign_state(call_vertex(1, 1),i)
                assign_state(call_vertex(2,2), j)
                assign_state(call_vertex(3,3), b)
                periodic_equalizer(vertices_list)

                origional_vertices_list = vertices_list.copy()
                for row_num, row in enumerate(origional_vertices_list):
                                for vert_num, vert in enumerate(row):
                                    if incomplete(vert):
                                        possible_states = []  
                                        for state in six_states:
                                            if align(vert, state):
                                                possible_states.append(state)

                                        org_vertices = [[copy.deepcopy(vertices_list)]*mu for mu in range(1,len(possible_states)+1)]


                                        for num, poss_state in  enumerate(possible_states):
                                            local_vert = org_vertices[num][0][row_num][vert_num]
                                            assign_state(local_vert, poss_state)
                                            periodic_equalizer(org_vertices[num][0])
                                            complete_states(org_vertices[num][0])
                                            if check_state(org_vertices[num][0]):
                                                counter+=1

                                        break

                                break


            
    return counter      

def empty(node):
    if node.right_state == "o" and node.left_state == 'o' and node.top_state =='o' and node.bottom_state =='o':
        return True
    return False





def fill_incomplete():
    global counter
    origional_vertices_list = vertices_list.copy()
    for row_num, row in enumerate(origional_vertices_list):
                    for vert_num, vert in enumerate(row):
                        if incomplete(vert):
                            possible_states = []  
                            for state in six_states:
                                if align(vert, state):
                                    possible_states.append(state)

                            org_vertices = [[copy.deepcopy(vertices_list)]*mu for mu in range(1,len(possible_states)+1)]


                            for num, poss_state in  enumerate(possible_states):
                                local_vert = org_vertices[num][0][row_num][vert_num]
                                assign_state(local_vert, poss_state)
                                periodic_equalizer(org_vertices[num][0])
                                complete_states(org_vertices[num][0])
                                if check_state(org_vertices[num][0]):
                                    counter+=1

                            break

                    break


##################
global first_empty
first_empty=[20, 20]
def fill_empty(vertices_list, i, j):
    global first_empty
    global counter
    if check_state(vertices_list):
                    counter+=1
            
    node = vertices_list[i][j]
    if sum(node.position) <= sum([L+1, L+1]):
        
        if empty(node):
            if node.position[0] < first_empty[0]:
                first_empty = node.position
            elif node.position[0] == first_empty[0]:
                if node.position[1] <= first_empty[1]:
                    node.position = first_empty
                
            for state in six_states:
                assign_state(node, state)
                periodic_equalizer(vertices_list)
                complete_states(vertices_list)
                periodic_equalizer(vertices_list)
                #
                #####count here
                
                
                #print(node.position)
                #print('\n')
                #
                if j ==L:  
                    j =-1
                    i+=1
                
                if sum(node.position) == sum([L+1, L+1]):
                    return
                else:
                    fill_empty(copy.deepcopy(vertices_list), i, j+1)
                
               

            #####
        
        if node.position != first_empty:

            if j ==L:
                    j =-1
                    i+=1
            if sum(node.position) == sum([L+1, L+1]):
                    return
            else:
                    fill_empty(copy.deepcopy(vertices_list), i, j+1)
    return

def operations():
    global counter
    periodic_equalizer(vertices_list)
    fill_empty(vertices_list, 0, 0)
    fill_incomplete()
    if check_state(vertices_list):
        counter+=1
```
